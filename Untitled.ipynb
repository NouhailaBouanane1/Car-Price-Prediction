```python
#Using quantilTransformer
#I will not apply that on Present_Price columns because it is highly correlated with Selling_Price
from sklearn.preprocessing import QuantileTransformer
qt=QuantileTransformer()

final_dataset['no_Year']=qt.fit_transform(final_dataset[['no_Year']])
#final_dataset['Selling_Price']=qt.fit_transform(final_dataset[['Selling_Price']])
final_dataset['Kms_Driven']=qt.fit_transform(final_dataset[['Kms_Driven']])
final_dataset['Owner']=qt.fit_transform(final_dataset[['Owner']])

print(skew(final_dataset['no_Year'],axis=0, bias=True))
print(skew(final_dataset['Selling_Price'],axis=0, bias=True))
print(skew(final_dataset['Kms_Driven'],axis=0, bias=True))
print(skew(final_dataset['Owner'],axis=0, bias=True))
```

```python
def plot_learning_curves(model, X_train, y_train,X_val,y_val):
    train_errors, val_errors = [], []
    for m in range(1, len(X_train) + 1):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    plt.plot(np.sqrt(train_errors), "r-+", linewidth=2, label="train")
    plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="test")
    plt.legend(loc="upper right", fontsize=14)
    plt.xlabel("Training set size", fontsize=14) 
    plt.ylabel("RMSE", fontsize=14) 
```

```python
plot_learning_curves(gradient_boosting_model,x_train,y_train,x_test,y_test)
```
